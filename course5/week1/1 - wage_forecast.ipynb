# Прогнозирование уровня средней заработной платы в России

```python
%pylab inline
import statsmodels.api as sm
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy import stats
import pandas as pd
import numpy as np
import warnings
from itertools import product
import matplotlib.pyplot as plt
```

### Загрузка и визуальное отображение ряда:

```python
wage = pd.read_csv('WAG_C_M.csv', index_col=['month'], header=0, names=['month','sums'], parse_dates=['month'], dayfirst=True)
wage.head(10)
```

```python
plt.figure(figsize(17, 7))
plt.xlabel('date')
plt.ylabel('wage, rub')
wage.sums.plot()
```

#### Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wage.sums).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.sums)[1])
```

Ряд нестационарен, на графиках декомпозиции виден сильный тренд и выраженная сезонность. Видно, что дисперсия растет со временем.

#### Проведем преобразование Бокса-Кокса для стабилизации дисперсии:

```python
wage['sums_stable'], lmbda = stats.boxcox(wage.sums)
#wage['sums_stable'] = stats.boxcox(wage.sums, lmbda=0)
print('Оптимальный параметр преобразования Бокса-Кокса lambda=%f' % lmbda)
```

```python
plt.figure(figsize(15,10))
wage.sums_stable.plot()
```

Видно, что разброс дисперсии уменьшился

#### Проведем сезонное дифференцирование

```python
wage['sums_year_diff'] = wage.sums_stable - wage.sums_stable.shift(12)

plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wage.sums_year_diff[12:]).plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wage.sums_year_diff[12:])[1])
```

Тренд не такой явный, критерий Дики-Фулера отвергает гипотезу о нестационарности.

Построим график автокорреляции:

```python
plt.figure(figsize(15,10))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wage.sums_year_diff[12:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

На графике автокорреляции виден явный убывающий тренд, значит полностью от тренда избавиться не удалось.

#### Добавим простое дифференцирование:

```python
wage['sums_diff'] = wage.sums_year_diff - wage.sums_year_diff.shift(1)

plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wage.sums_diff[13:]).plot()
print("Критерий Дики-Фуллера: p=%g" % sm.tsa.stattools.adfuller(wage.sums_diff[13:])[1])
```

Кажется, тренда больше нет, критерий Дики-Фулера теперь еще более уверенно отвергает гипотезу о нестационарности.

### Подбор модели

#### Построим графики автокорреляции и частичной автокорреляции:

```python
plt.figure(figsize(15,10))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wage.sums_diff[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(wage.sums_diff[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

Теперь на графике автокорреляции тренда не видно.

Значимого сезонного лага автокорреляции нет, последний значимый лаг автокорреляции - 1. 
На графике частичной автокорреляции: последний значимый сезонный лаг - 12, последний значимый лаг (должен быть меньше сезонного, иначе модель будет неоднозначно определена) - 1.

```python
Q = 0
q = 1
P = 1
p = 1
d = 1
D = 1

qs = range(0, q+1)
Qs = range(0, Q+1)
Ps = range(0, P+1)
ps = range(0, p+1)
```

```python
parameters = product(qs, Qs, Ps, ps)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for (qi, Qi, Pi, pi) in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(wage.sums_stable, order=(pi, d, qi), 
                                        seasonal_order=(Pi, D, Qi, 12)).fit(disp=-1)
    except ValueError:
        print ('Wrong parameters: q=%i, Q=%i, P=%i, p=%i' % (qi, Qi, Pi, pi))               
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = (qi, Qi, Pi, pi) 
    results.append([(qi, Qi, Pi, pi), model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

#### Лучшая модель:

```python
print(best_model.summary())
```

#### Остатки лучшей модели:

```python
resid = best_model.resid[13:]
plt.figure(figsize(15,8))
plt.subplot(211)
resid.plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(resid, 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(resid)[1])
```

Остатки стационарны (подтверждается критерием Дики-Фуллера), неавтокоррелированы (подтверждается коррелограммой).
Однако присутствует смещение (критерий Стьюдента не отвергает гипотезу о нормальном распределении).

Проверим остатки на автокоррелированность критерием Льюнга-Бокса.

$H_0$: остатки являются случайными (то есть представляют собой белый шум).

$H_1$: $H_0$ неверна.

```python
(lbvlue, pvalue) = acorr_ljungbox(resid)
print pvalue
print np.where(pvalue > 0.05)[0]
```

Видим, что остатки неавтокоррелированы (гипотеза $H_0$ не отвергается для 39 лагов).

Так как остатки немного смещены, скорректируем прогноз на константу:

```python
resid_shift = np.mean(resid)
print resid_shift
```

### Сравнение модели с данными:

```python
def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
wage['model'] = invboxcox(best_model.fittedvalues, lmbda) - resid_shift
plt.figure(figsize(25,15))
wage.sums.plot()
wage.model[13:].plot(color='r')
plt.ylabel('Wage, rub')
pylab.show()
```

```python
n = len(best_model.fittedvalues)
```

Видим, что модель достаточно точная.
